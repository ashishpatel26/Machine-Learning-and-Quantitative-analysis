```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import quandl
import pandas_datareader.data as web
from datetime import datetime


%matplotlib inline
```

```python
my_year = 2017
my_month = 1
my_day = 2
my_hour = 13
my_minute = 30
my_second = 15
```

```python
start = datetime(2017, 9, 1)
end = datetime(2017, 11, 24)
```

```python
ryb = web.DataReader('RYB', 'google', start, end)
```

```python
ryb.head()
```

```python
ryb['Close'].plot(figsize = (12,5), grid = True)
```

```python
idx = ryb.index
idx
```

```python
close_price = ryb['Close'].values
close_price
```

```python
import matplotlib.dates as dates

fig, ax = plt.subplots(figsize = (12, 8))
ax.plot_date(idx, close_price, 'r-')

ax.yaxis.grid(True)
ax.xaxis.grid(True)

# Location
ax.xaxis.set_major_locator(dates.MonthLocator())
# Formating
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

fig.autofmt_xdate()
plt.tight_layout()
```
