```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('ggplot')
```

```python
df = pd.read_csv('walmart_stock.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
# Average moving value, 7 days,  Moving average
df.rolling(7).mean().head(14)
```

```python
df['Open'].rolling(7).mean().plot(figsize=(16,6))
```

```python
df['Open'].plot(c = 'r', label = 'open')
# Weekly moving average
df.rolling(window = 7).mean()['Close'].plot(figsize=(18,9), c = 'g', label = 'Week')

# Monthly moving average
df.rolling(window = 30).mean()['Close'].plot(figsize=(18,9), c = 'purple', label = 'Month')

plt.legend(loc = 0)
```

```python
df['Close 30 Day MA'] = df['Close'].rolling(window = 30).mean()
df[['Close 30 Day MA', 'Close']].plot(figsize=(16,6))
plt.tight_layout()
```

```python
# Everyday mean valus
df['Close'].expanding().mean().plot(figsize=(16,6))
```

# Bollinger Bands

```python
plt.style.context(('dark_background'))

# style selection:http://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html
plt.style.use('bmh')
              
# Close 20 MA
df['Close: 20 Day Mean'] = df['Close'].rolling(window=20).mean()
# Upper = 20MA + 2 * std(20)
df['Std: 20 Days'] = df['Close'].rolling(window=20).std()
df['Upper'] = df['Close: 20 Day Mean'] + 2 * df['Std: 20 Days']
# Lower = 20MA - 2 * std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2 * df['Std: 20 Days']
# Close

# Alternative way
# df[['Close: 20 Day Mean', 'Upper', 'Lower', 'Close']].tail(200).plot(figsize = (16,9))
df[['Close: 20 Day Mean', 'Upper', 'Lower', 'Close']].plot(figsize = (15,7))
```
