```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')

%matplotlib inline
```

```python
x = np.linspace(0,10,1000)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y,)
plt.xlabel('Time')
plt.ylabel('Some function of time')
plt.title('My cool chart')
plt.show()
```
